```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('./b1abb33c4f.csv')
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

for var in variaveis:
    plt.figure(figsize=(8, 4))
    df[var].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Frequência da variável {var}')
    plt.xlabel(var)
    plt.ylabel('Frequência')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apgar5_mean = df.groupby('CONSULTAS')['APGAR5'].mean()


plt.figure(figsize=(8, 4))
apgar5_mean.plot(kind='bar', color='skyblue')
plt.title('Média do APGAR5 por número de Consultas')
plt.xlabel('Número de Consultas')
plt.ylabel('APGAR5 Médio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
variaveis = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']
eixos = [(0, 0), (0, 1), (1, 0), (1, 1)]
fig, axs = plt.subplots(2, 2, figsize=(14, 10))


for var, (i, j) in zip(variaveis, eixos):
    df[var].value_counts().plot(kind='bar', ax=axs[i, j], color='skyblue')
    axs[i, j].set_title(f'Frequência da variável {var}')
    axs[i, j].set_xlabel(var)
    axs[i, j].set_ylabel('Frequência')
    axs[i, j].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
